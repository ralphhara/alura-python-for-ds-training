##### Detectando ***encoding*** de um arquivo CSV
```
import chardet

with open('/content/dados.csv', 'rb') as file:
    print(chardet.detect(file.read()))
```

Note que a primeira linha deste cÃ³digo abre o arquivo CSV em modo de leitura binÃ¡ria rb e atribui o objeto de arquivo retornado a uma variÃ¡vel chamada file.

Na segunda linha do cÃ³digo, o conteÃºdo do arquivo CSV Ã© lido usando o mÃ©todo read() e o resultado Ã© passado para a funÃ§Ã£o chardet.detect(), que retorna um dicionÃ¡rio contendo informaÃ§Ãµes sobre a codificaÃ§Ã£o mais provÃ¡vel do arquivo. O resultado Ã© impresso com a funÃ§Ã£o print() que mostra a codificaÃ§Ã£o e a confianÃ§a associada a essa codificaÃ§Ã£o.

Ao executar o cÃ³digo Giovanna obteve o seguinte resultado:

```
{'encoding': 'ISO-8859-1', 'confidence': 0.73, 'language': ''}
```

Para especificar a codificaÃ§Ã£o correta ao carregar o arquivo CSV com a biblioteca Pandas, Ã© possÃ­vel usar um parÃ¢metro chamado `encoding`:

`df = pd.read_csv('/content/dados.csv', encoding='ISO-8859-1')`

````python
from sqlalchemy import create_engine, MetaData, table, inspect

# create_engine: serve para criar o motor, ou seja, a conexÃ£o com o banco de dados que o farÃ¡ funcionar.
# MetaData: Essa Ã© uma classe que representa os meta dados do banco de dados, isso inclui informaÃ§Ãµes sobre as tabelas, colunas e outros objetos.
# Table: representa uma tabela em um banco de dados e Ã© usada para manipular dados em uma tabela especÃ­fica.
# inspect: permite inspecionar a estrutura do banco de dados. Em seguida, executamos a cÃ©lula.
````

````python
engine = create_engine('sqlite:///:memory:')
````

````python
url_banco = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
````

````python
dados_banco = pd.read_csv(url_banco)
````

````python
dados_banco.head()
````

````python
dados_banco.to_sql('clientes', engine, index=False)
````

````python
inspector = inspect(engine)
````
